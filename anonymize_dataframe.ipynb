```python
import pandas as pd
import numpy as np
```

```python
df_path = "./RAW_DATA_REFINED.kin"
full_df  = pd.read_pickle(df_path)
full_df
```

```python
def list_unique_elements_in_df_column(df, column_name):
    unique_elements = []
    for index, row in df.iterrows():
        element = row[column_name]
        if element not in unique_elements:
            unique_elements.append(element)
    return unique_elements
```

```python
studies = sorted([x for x in list_unique_elements_in_df_column(full_df,"Study")])

study_anon_dict = {
    "Actiyas": "Study_1",
    "GMK" : "Study_5",
    "Lima" : "Study_6",
    "Toshi" : "Study_7"
}
```

```python
# rename study names in dataframe using .apply
full_df["Study"] = full_df["Study"].apply(lambda x: study_anon_dict[x])
```

```python
study1_df = full_df[full_df["Study"] == "Study_1"]
print(list_unique_elements_in_df_column(study1_df,"Patient"))

study_1_patient_anon_dict = {
    "T01L" : "Patient_1",
    "T01R" : "Patient_2",
    "T02L" : "Patient_3",
    "T02R" : "Patient_4",
    "T03L" : "Patient_5",
    "T03R" : "Patient_6"      
}

# rename only specific portions of dataframe using .replace
full_df[full_df["Study"] == "Study_1"] = full_df[full_df["Study"] == "Study_1"].replace({"Patient": study_1_patient_anon_dict})
```

```python
# now repeat this for the other studies
study3_df = full_df[full_df["Study"] == "Study_5"]
print(list_unique_elements_in_df_column(study3_df,"Patient"))
```

```python
study6_df = full_df[full_df["Study"] == "Study_6"]
print(list_unique_elements_in_df_column(study6_df,"Patient"))

study6_anon_dict = {
    'Patient 77-08-MB' : 'Patient_3',
    'Patient 77-11-RW' : 'Patient_6'}

full_df[full_df["Study"] == "Study_6"] = full_df[full_df["Study"] == "Study_6"].replace({"Patient": study6_anon_dict})
```
